1.FIBONOCCI SERIES

```python
N=int(input(" how many terms u want to print fibonacci series:"))
n1,n2=0,1
sum=0
count=1
print("fibonocci series",end="")
while count<=N:
  print(sum,end="")
  count+=1
  n1=n2
  n2=sum
  sum=n1+n2
```

2.CHECK GIVEN NUMBER IS PRIME NUMBER OR NOT

```python
n=int(input("enter a number:"))
if n>1:
  for i in range(2,n):
    if (n%i)==0:
      print(n,"this is not prime number")
      break
  else:
      print(n,"this is prime number")
```

10.PRINTING PATTERN

PYRAMID PATTERN


```python
n=int(input("Enter a rows:"))
for i in range(0,n):
  for j in range(0,i+1):
    print("*",end=" ")
  print()
      
```

HALF PYRAMID REVERSE

```python
n=int(input("enter a rows:"))
for i in range(n+1,0,-1):
  for j in range(0,i-1):
    print("*",end="")
  print()
```

TRIANGLE PYRAMID

```python
n = int(input("enter a rows: "))
m = (2 * n) - 2
for i in range(0, n):
    for j in range(0 ,m):
       print(end=" ")
    m = m - 1
    for j in range(0, i + 1):
      print("* ", end=' ')
    print(" ")
```
